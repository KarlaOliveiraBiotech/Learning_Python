The better approach is using one square bracket for each level we're descending and it's important to note we can also have nested dicitonaries. 
<br><br> In the code below, we want to get value associated with key color. 
<br>See the level descending idea as it follows:

```python
info = {'personal_data': 
        {'name': 'Lauren', 
         'age': 20,
         'major': 'Information Science', 
         'physical_features': 
         {'color': {'eye': 'blue', 
                    'hair': 'brown'}, 
                    'height': "5'8"
        }, 
        'other': 
            {'favorite_colors': ['purple', 'green', 'blue'], 
             'interested_in': ['social media', 'intellectual property', 'copyright', 'music', 'books']
             }
        }
}

print("First level of descending")
color = info['personal_data'] 
print(color, "\n")

print("Second level of descending")
color = info['personal_data']['physical_features']
print(color, "\n")

print("Third level of descending")
color = info['personal_data']['physical_features']['color']
print(color)
```

```python
import json

def pretty(obj):
    return json.dumps(obj, sort_keys = True, indent = 2)

as_string = pretty(info)
print(as_string)
```

For nested iteration we can use for loops such as follows:

```python
nested1 = [['a', 'b', 'c'], ['d', 'e'], ['f', 'g', 'h']]

for outer in nested1:
    print("level 1:  ", outer)
    for inner in outer:
        print("  level 2:  ", inner)
```

Below, we want to save every string that contains 'b' into  a list called 'b_strings'.

```python
groceries = [['apples', 'bananas', 'oranges', 'blueberries', 'lemons'], 
             ['carrots', 'peas', 'cuccumber', 'green beans'], 
             ['root beer', 'cranberry juice']]


# This is the easy way to do
b_strings = []
for lists in groceries: 
    for word in lists:
        if 'b' in word:
            b_strings.append(word)
print("This was done by the easy way: ", b_strings)


# This is the hard way to do
new_b_strings = []
for lists in groceries: 
    for word in lists:
        for char in word:
            if char == 'b':
                new_b_strings.append(word)
                break
print("This was done by the hard way: ", new_b_strings)

        
```

Shallow version: same inner list is included to multiple outer lists.

```python
original = [['dogs', 'puppies'], ['cats', 'kittens']]
copied = original[:]

print(copied is original)
print('---')
print(copied == original, "\n")

original[0].append(['canines'])

print("This is original appended: ", original)
print("And this is the copied version: ", copied) 
```

If we want to make a copy that is completely independent of its original, we must do a **deep copy**, which means doing element-by-element' copy such as it follows:

```python
original = [['dogs', 'puppies'], ['cats', 'kittens']]
outer_copy = []

for inner_list in original:
    inner_copy = []
    for item in inner_list:
        inner_copy.append(item)
    outer_copy.append(inner_copy)


original[0].append(['canines'])


print(outer_copy is original)
print('---')
print(outer_copy == original, "\n")

print("This is original appended: ", original, "\n")
print("And this is the copied version: ", outer_copy) 
```

The previous version does an accumulation during deep copy. But it can be ripped off in the following mode: 

```python
original = [['dogs', 'puppies'], ['cats', 'kittens']]
outer_copy = []

for inner_list in original:
    inner_copy = inner_list[:]
    outer_copy.append(inner_copy)

original[0].append(['canines'])


print(outer_copy is original)
print('---')
print(outer_copy == original, "\n")

print("This is original appended: ", original, "\n")
print("And this is the copied version: ", outer_copy) 
```

A third and easier way to make deep copy is very useful when we don't know the levels of lists and the structure is much more complex. 
<br>In this case, we import a module called "copy" that has a built-in function that does it. It is based on recursion. See below how to do it.

```python
import copy
original = [['canines', ['dogs', 'puppies']], ['felines', ['cats', 'kittens']]]

shallow_copy = original[:]
deep_copy = copy.deepcopy(original)

original.append("Hi there")
original[0].append(['marsupials'])

print("------ Original ------")
print(original)

print("------ Deep copy ------")
print(deep_copy)

print("------ Shallow copy ------")
print(shallow_copy)
```

```python
print(type(res))
print(res.keys())
res2 = res['statuses']
print("----Level 2-----")
print(type(res2)) # it's a list!
print(len(res2))
```

2. Level 3

```python
import json
print(type(res))
print(res.keys())
res2 = res['statuses']
print("----Level 2: a list of tweets-----")
print(type(res2)) # it's a list!
print(len(res2))  # looks like one item representing each of the three tweets
for res3 in res2[:1]:
   print("----Level 3: a tweet----")
   print(json.dumps(res3, indent=2)[:30])
   print(type(res3)) # it's a dictionary
   print(res3.keys())
```

3. Level 4

```python
import json
print(type(res))
print(res.keys())
res2 = res['statuses']
print("----Level 2: a list of tweets-----")
print(type(res2)) # it's a list!
print(len(res2))  # looks like one item representing each of the three tweets
for res3 in res2[:1]:
   print("----Level 3: a tweet----")
   print(json.dumps(res3, indent=2)[:30])
   res4 = res3['user']
   print("----Level 4: the user who wrote the tweet----")
   print(type(res4)) # it's a dictionary
   print(res4.keys())
   print(res4['screen_name'], res4['created_at'])
```
